```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
img_size = (150, 150) 
a = np.random.rand(np.prod(img_size)).reshape(img_size)
```

```python
def center(img):
    return np.array(img.shape) // 2

def rotate(x, y, ang):
    dx = x*np.cos(ang) - y*np.sin(ang)
    dy = x*np.sin(ang) + y*np.cos(ang)
    
    return dx, dy
    
cen_x, cen_y = center(a)

fig, ax = plt.subplots(1,1)

ax.imshow(a)
ax.plot(cen_x, cen_y, 'ro')

off_x = np.zeros(10)
off_y = np.arange(10)*5

ax.plot(off_x+cen_x, off_y+cen_y, 'bo')

for i in np.arange(45, 355, 45):
    dx, dy = rotate(off_x, off_y, np.deg2rad(i))
    ax.plot(dx+cen_x, dy+cen_y, 'ro')

fig.show()
```
